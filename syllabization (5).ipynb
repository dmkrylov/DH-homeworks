```python
import unicodedata

yihhi = 'ὕἶῃὴΐ'
xyhta = unicodedata.normalize('NFD',yihhi)
rough = xyhta[1]
smooth = xyhta[4]
grave = xyhta[9]
acute = xyhta[2]
circumflex = xyhta[5]
iota_subscript = xyhta[7]
diaeresa = xyhta[11]
diacritics = rough+smooth+grave+acute+circumflex+iota_subscript+diaeresa

vowels = 'ωηοεαυι'
consonants = 'ςρτθπσδφγκλχβνμ'
doubles = 'ζξψ'

muta = 'τθπδφγκχβ'
liquida = 'ρλνμ'
dyphtongs = ['οι','ει','αι','υι','ου','ευ','αυ','ηυ']
punct = "!#$%&()*+,-.:;<=>?^_~„“«»†*—-"

def syllabization (s):
    syllables = 0
    s = unicodedata.normalize('NFD', s)
    for symbol in s:
        if symbol in vowels:
            syllables +=1
    for i in range (len(s)-1):
        if s[i]+s[i+1] in dyphtongs:
            syllables -=1
    for j in range (len(s)-2):
        if s[j]+s[j+1] in dyphtongs and s[j+2]==diaeresa:
            syllables +=1
    return syllables


def divide_by_antisyllables (s):
    s = unicodedata.normalize('NFD', s)
    s= s.lower()
    s+='</l>'
    i=0
    while i<len(s)-2:
        if s[i]+s[i+1] in dyphtongs and s[i-1]!='@':
            s = s[:i]+'@'+s[i:]    
        elif s[i] in vowels and s[i-1]!='@' and s[i-1]+s[i] not in dyphtongs:
            s = s[:i]+'@'+s[i:]
        elif s[i] in vowels and s[i-1]!='@' and s[i+1] == diaeresa:
            s = s[:i]+'@'+s[i:]
        i+=1    
    syllables = s.split('@')
    #print(syllables)
    return syllables
#print(divide_by_antisyllables('<l n="176">παῖδες, φρονεῖν χρή·  ξὺν φρονοῦντι δ’ ἥκετε</l>'))
def define_length(AS):
    m='x'
    AS = unicodedata.normalize('NFD', AS)
    local_consonants=[]
    local_vowels=[]
    for symbol in AS:
        if symbol not in (vowels+consonants+doubles+diacritics):
            AS = AS.replace(symbol,'')
        if symbol in doubles or symbol in vowels[:2] or symbol == iota_subscript or symbol == circumflex:
            m='l'
            break
        if symbol in consonants:
            local_consonants.append(symbol)
        if symbol in vowels:
            local_vowels.append(symbol)
 #   print(AS)
    if m=='l':
        pass
    elif len(local_vowels)>1:
        m='l'
    elif len(local_consonants)>2:
        m='l'
    elif len(local_consonants)==2:
        if local_consonants[0] in muta and local_consonants[1] in liquida:
            pass
        else:
            m='l'
    elif local_vowels[0] in vowels[2:4]:
        m='s'
    return m

def metr_config(s):
    list_of_AS = divide_by_antisyllables(s)[1:]
    config=''
    for AS in list_of_AS:
        config+=define_length(AS)
    #print(config)
    return config

#print(metr_config(', <l>πέποιθα &lt;δὴ&gt;  τὸν Διὸς ἀντίτυπον ἔχοντ’'))
```
